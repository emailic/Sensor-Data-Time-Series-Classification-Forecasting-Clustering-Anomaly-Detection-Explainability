# Data Preparation

## - Correzione Secondi

l'attributo 'date' presenta degli istanti in cui viene rilevato il 59-esimo secondo invece che il minuto intero.

Noi quÃ¬ correggiamo questa cosa riportando tutti gli istanti ad essere divisi da 1 minuto intero, aggiundendo 1 secondo alle rilevazioni in corrispondenza del 59-esimo secondo per farle arrivare al minuto tondo.

#### Training

```python
df, ts = df_handle.import_df(file_name='datatraining.txt', search_in_folder=files_path,
                             date_col_name='date', ts_name='Temperature', time_freq='min')
```

```python
df
```

```python
import pandas as pd
from pandas.tseries.offsets import DateOffset
```

```python
dates_series = pd.Series(df.index)
dates_series
```

```python
i=0
for date in dates_series:
    if date.second == 59:
        dates_series[i] = date + DateOffset(seconds=1)
    i+=1
```

```python
dates_series
```

```python
df.index = dates_series
```

```python
del df['year'], df['month'], df['day'], df['hour'], df['minute'], df['second']
```

```python
df
```

```python
df.to_csv(path_or_buf = files_path + '\\' + 'training.csv', sep=',', encoding='utf-8', index=True)
```

## - Unione dei due test con in Training per formare un unico grosso DataFrame

il primo test set (test1) si riferisce ad un momento temporale precedente al training quindi viene aggiunto prima di tutti, poi viene aggiunto il training set e poi il secondo test set (test2) che si riferisce ad un momento temporale successivo al training.

```python
df1 = df_handle.import_df(file_name='test1.csv', search_in_folder=files_path, date_col_name='date')
df1
```

```python
df2 = df_handle.import_df(file_name='training.csv', search_in_folder=files_path, date_col_name='date')
df2
```

```python
df3 = df_handle.import_df(file_name='test2.csv', search_in_folder=files_path, date_col_name='date')
df3
```
